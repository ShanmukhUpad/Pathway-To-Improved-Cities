Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

Load Datasets

```python
bike_routes = pd.read_csv("transport_access_raw_data/Bike_Routes.csv")
bus_stops = pd.read_csv("transport_access_raw_data/Bus_Stops.csv")
divvy_bicycle = pd.read_csv("transport_access_raw_data/Divvy_Bicycle_Stations.csv")

crimes = pd.read_csv("safety_raw_data/Crimes_2026.csv")
fire_stations = pd.read_csv("safety_raw_data/Fire_Stations.csv")
police_stations = pd.read_csv("safety_raw_data/Police_Stations.csv")
```

Handle Missing Data

```python
def clean_missing(df):

    # remove duplicates
    df = df.drop_duplicates()

    # drop rows missing coordinates
    if 'latitude' in df.columns:
        df = df.dropna(subset=['latitude','longitude'])

    # fill other missing values
    df = df.fillna("Unknown")

    return df
```

```python
bike_routes = clean_missing(bike_routes)
bus_stops = clean_missing(bus_stops)
divvy_bicycle = clean_missing(divvy_bicycle)
crimes = clean_missing(crimes)
fire_stations = clean_missing(fire_stations)
police_stations = clean_missing(police_stations)
```

Standardize Column Names

```python
def standardize_columns(df):
    df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")
    return df

bike_routes = standardize_columns(bike_routes)
bus_stops = standardize_columns(bus_stops)
divvy_bicycle = standardize_columns(divvy_bicycle)
crimes = standardize_columns(crimes)
fire_stations = standardize_columns(fire_stations)
police_stations = standardize_columns(police_stations)
```
