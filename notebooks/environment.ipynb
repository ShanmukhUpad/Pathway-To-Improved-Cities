```python
import pandas as pd
energy = pd.read_csv('raw_data/Energy_Usage_2010_20260421.csv')
benchmark = pd.read_csv('raw_data/Chicago_Energy_Benchmarking_-_Covered_Buildings_20260421.csv')
complaints = pd.read_csv('raw_data/CDPH_Environmental_Complaints_20260421.csv')

energy.head(), benchmark.head(), complaints.head()
```

Clean Energy Usage

```python

energy.columns = energy.columns.str.strip().str.replace(' ', '_')

# convert numeric columns
numeric_cols = energy.select_dtypes(include='object').columns
for col in numeric_cols:
    energy[col] = pd.to_numeric(energy[col], errors='ignore')

energy = energy.drop_duplicates()
energy = energy.fillna(0)

energy.head()
```

Clean Benchmark Data

```python

benchmark.columns = benchmark.columns.str.strip().str.replace(' ', '_')
benchmark = benchmark.drop_duplicates()
benchmark = benchmark.fillna(method='ffill')

benchmark.head()
```

Clean Complaints Data

```python

complaints.columns = complaints.columns.str.strip().str.replace(' ', '_')
complaints['COMPLAINT_DATE'] = pd.to_datetime(complaints['COMPLAINT_DATE'], errors='coerce')
complaints = complaints.drop_duplicates()
complaints = complaints.fillna('Unknown')

complaints.head()
```

Save Cleaned Data

```python
energy.to_csv('clean_data/clean_energy.csv', index=False)
benchmark.to_csv('clean_data/clean_benchmark.csv', index=False)
complaints.to_csv('clean_data/clean_complaints.csv', index=False)
```
