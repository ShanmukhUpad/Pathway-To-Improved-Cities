### Grouping / Feature Engineering Phase

```python
df['Measurement Description'].value_counts()
```

```python
cumulative_precipitation = df[df['Measurement Type'] == "CumulativePrecipitation"]
cumulative_precipitation
```

#### Visualization of cumulative precipation over time for each of the 3 locations

```python
cumulative_precipitation['Measurement Time'] = pd.to_datetime(cumulative_precipitation['Measurement Time'])
cumulative_precipitation = cumulative_precipitation.sort_values(by='Measurement Time')
cumulative_precipitation['Location'].unique()
locations = cumulative_precipitation['Location'].unique()[:3]
```

```python
cumulative_precipitation = cumulative_precipitation.groupby(
    ['Measurement Time', 'Location']
)['Measurement Value'].mean().reset_index()
cumulative_precipitation
```

```python
plt.figure(figsize=(10, 6))

for loc in locations:
    subset = cumulative_precipitation[cumulative_precipitation['Location'] == loc]
    
    plt.plot(
        subset['Measurement Time'],
        subset['Measurement Value'],
        label=f'Location {loc}'
    )

plt.xlabel('Measurement Time')
plt.ylabel('Cumulative Precipitation')
plt.title('Cumulative Precipitation Over Time by Location')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
pivot_df = cumulative_precipitation.pivot_table(
    index='Measurement Time',
    columns='Location',
    values='Measurement Value'
)
pivot_df = pivot_df.interpolate()
pivot_df.plot(figsize=(10, 6))

plt.xlabel('Measurement Time')
plt.ylabel('Cumulative Precipitation')
plt.title('Cumulative Precipitation Over Time by Location')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
